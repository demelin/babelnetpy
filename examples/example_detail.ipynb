```python
from babelnetpy.babelnet import BabelNet

bn = BabelNet(open("key.txt", "r").read())

Ids = bn.getSynset_Ids("correr", "ES")
print(Ids[0])
```

## search language is memorized in the class

```python
bn.lang
```

```python
bn.getSynsets(Ids[0].id)
```

```python
# don't change first searchlang
bn.lang
```

## getSynsets has lang parameter and can translate

```python
# when you want to change lang, you use change_lang parameter
bn.getSynsets(Ids[0].id, ["EN", "ES"], change_lang=True)
```

```python
# change first searchlang
bn.lang
```
